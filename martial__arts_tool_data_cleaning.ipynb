```python
import pandas as pd
import numpy as np
```

```python
data_ = pd.read_csv("threepagesdata.csv", encoding = 'utf8')
data_.head()
```

```python
print("5 hang cuoi cung cua khung du lieu\n")
data_.tail(10)
```

```python
#Hiển thị các tên cột trong Dataframe
print(data_.columns)
```

```python
#xử lí cột percent_response_rate
dataa = data_.replace('Không đủ thông tin', np.NaN) #đổi các giá trị thành NA
```

```python
#Xóa các giá trị NA trong dataframe
data2 = dataa.fillna(method='ffill') #tăng giá trị lên 1
data3 = data2.fillna('0') #chèn 0 vào các giá trị Na
data3.head(5)
```

```python
#Xóa dấu % trong dataframe
data3['percent_positive_reviews'] = (data3['percent_positive_reviews'].str.strip('%').astype(int))
data3['percent_response_rate'] = (data3['percent_response_rate'].str.strip('%').astype(int))
data3['percentage_of_on_time_delivery'] = (data3['percentage_of_on_time_delivery'].str.strip('%').astype(int))
data3['sale'] = (data3['sale'].str.strip('%').astype(str))
data3.head(5)
```

```python
#Đổi toàn bộ giá trị 'Miễn phí' trong dataframe
data4 = data3.replace(to_replace ="Miễn phí", 
                            value ="0")
```

```python
#xử lí những giá trị trong cột có '₫'
data4['delivery_charges'] = (data4['delivery_charges'].str.strip('₫').astype(str))
data4['initial_price'] = (data4['initial_price'].str.strip('₫').astype(str))
data4['sale_price'] = (data4['sale_price'].str.strip('₫').astype(str))
```

```python
#Xử lí những dấu '-' trong cột sale
data4['sale'] = (data4['sale'].str.strip('-').astype(int))
data4.head()
```

```python
#Đổi giá trị trong cột lie
data4["lie"].replace({"7 ngày trả hàng dễ dàng": "7 ngày trả hàng", "7 ngày trả hàng cho Nhà bán hàng": "3 ngày trả hàng", 
                      "0":"Không được trả hàng"}, inplace=True)
data4.head(5)
```

```python
#Kiểm tra định dạng dataframe
data4.dtypes
```

```python
#Lưu thành file csv
data4.to_csv('newthreepagesdata.csv', index=False)
```
